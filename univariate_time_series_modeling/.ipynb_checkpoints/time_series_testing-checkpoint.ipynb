```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,3)

from google.cloud import bigquery
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
from statsmodels.tsa.seasonal import STL
from scipy import stats
#from statsmodels.tsa.api import VAR
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tools.eval_measures import rmse, aic
import datetime
import re
import pandas as pd
import itertools

from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf

import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.stattools import adfuller
from itertools import product

import dateutil

import random

from random import randint

from IPython.display import clear_output
import matplotlib.pyplot as plt

def get_diff(arr, verbose = False):

        """
    Uses an Adjusted Dickey Fuller test to determine if first or second differencing induces stationarity.  Returns minimum difference.    

    Parameters:
    -----------
        df : dataframe - in time series format.  Must have dates as index.  Must be univariate (i.e. one column)

    Returns:
    -------
        difference : int - minimum difference needed to induce stationarity.

        """
        #arr = arr.dropna()

        difference = 0

        pvalue = adfuller(arr)[1]

        if verbose == True:

            print(f'diff-{difference} : pvalue-{pvalue}')

        while pvalue >= .05:

            difference += 1

            arr = arr.diff().fillna(0)

            pvalue = adfuller(arr)[1]

            if verbose == True:

                print(f'diff-{difference} : pvalue-{pvalue}')

            if difference == 10:

                difference = 0

                break

        return(difference)

def get_min_ma(arr):

        """
    Parameters:
    -----------
        df : dataframe - in time series format.  Must have dates as index.  Must be univariate (i.e. one column)

    Returns:
    -------
        ma : int - minimum lag suggested for MA model.  

        """


        from statsmodels.tsa.stattools import pacf

        pacf_len = np.floor(arr.shape[0]/2) -1

        pk = pacf(arr, nlags=pacf_len, alpha = .05)

        #pk[0] is the correlation coefficient, regardless of acf or pacf
        #pk[1] is the confidence interval given by alpha
        #pk[1][:,0] is the low end, pk[1][:,1] is high
        #pk[1][:,{0 or 1}] - pk[0] allows us to center the confidence interval around zero.  
        #If the corr coef is outside of this shifted bound, it is likely not significant

        corr_df = pd.DataFrame({\
                      'corr_coeff':pk[0]\
                      ,'ci_low':pk[1][:,0] - pk[0]\
                      ,'ci_hi':pk[1][:,1] - pk[0]})

        #Grabs a list of all significant lags
        vals = corr_df.loc[corr_df.corr_coeff.between(corr_df.ci_low, corr_df.ci_hi)==False].reset_index()['index'].values


        #Finds the minimum significant lag.  Zips vals against a list of len = .5*len -1.  Then filters them if the pairs arent x[0]==x[1].
        if len(vals) == 2 and vals[1] == 1:
            ma = 1
        else:
            #ma = min(min([v for v in zip(vals,vals[1:]) if v[0]+1 != v[1]]))
            ma = max(max([x for x in zip(vals,range(0,int(pacf_len))) if x[0] == x[1]]))

        return(ma)

def auto_sarimax_v2(series,verbose=False):   


    print("Figuring out Difference")
    difference = get_diff(series)
    print("Figuring out Moving Average")
    ma = get_min_ma(series)
    p,d,q = range(0,5),[difference],range(ma,5)
    
    df = pd.DataFrame({'date':pd.date_range(datetime.date.today(), freq='M', periods=len(series))\
                       ,'data':series})\
                        .set_index('date')

    print("Figuring out Seasonality")
    seasonal_result = seasonal_decompose(df, model='additive',extrapolate_trend='freq').seasonal
    print("Figuring out Seasonal Difference")
    seasonal_difference = get_diff(seasonal_result)
    print("Figuring out Seasonal Moving Average")
    seasonal_ma = get_min_ma(seasonal_result)
    P,D,Q,s = range(0,5),[seasonal_difference],range(seasonal_ma,5),[12]

    pdq = list(itertools.product(p, d, q))
    seasonal_pdq = list(itertools.product(P, D, Q, s))
    all_param = list(itertools.product(pdq,seasonal_pdq))
    print(all_param)
    pdq_df_sarimax = pd.DataFrame(columns = ['order','aic','bic','mse'])

    for i,param in enumerate(all_param):
        if verbose == True:
            clear_output(wait=True)
            print(f'At {i/len(all_param):.2%} Modeling {param}...')

        try:
            print("Modeling...")
            model = SARIMAX(series
                        ,order=param[0]
                        ,seasonal_order=param[1]
                       ,enforce_stationarity=False
                        ,enforce_invertibility=False
                       ,simple_differencing=True)
            results = model.fit(disp=0)
            fore = results.get_forecast(steps=6)
            f = fore.summary_frame()
            pdq_df_sarimax = pdq_df_sarimax.append(\
                                           pd.DataFrame(data=\
                                                        {'order':[param]\
                                                         ,'aic':[results.aic]\
                                                         ,'bic':[results.bic]\
                                                         ,'mse':[results.mse]\
                                                         ,'series':[series]\
                                                         ,'forecast':[f['mean'].to_list()]\
                                                         , 'mean_se' : [f['mean_se'].to_list()]\
                                                         , 'mean_ci_lower' : [f['mean_ci_lower'].to_list()]\
                                                         , 'mean_ci_upper' : [f['mean_ci_upper'].to_list()]\
                                                        }))
            
            #mean_se  mean_ci_lower  mean_ci_upper
        except:
            continue


    return(pdq_df_sarimax)
```

```python
def on_screen_number_one(df,l,ax_=None):
    # append a pandas series to an existing pandas series

    size = len(l)
    
    order = df[df.mse==min(df.mse)].order.to_list()[0]
    aic = df[df.mse==min(df.mse)].aic.to_list()[0]
    bic = df[df.mse==min(df.mse)].bic.to_list()[0]
    mse = df[df.mse==min(df.mse)].mse.to_list()[0]
    point = df[df.mse==min(df.mse)].forecast.to_list()[0][0]
    #forecast = df[df.mse==min(df.mse)].forecast

    if ax is not None:
        plt.plot(l.append(pd.Series(df[df.mse==min(df.mse)].forecast.to_list()[0])).reset_index(drop=True),ax=axs[ax_])
    else:
        plt.plot(l.append(pd.Series(df[df.mse==min(df.mse)].forecast.to_list()[0])).reset_index(drop=True))
    plt.axvline(x=size, color='r')
    plt.title(f'order: {order}\nmse: {mse}\n aic: {aic}\n bic: {bic}')
    plt.annotate(point, (size+1,point))
    plt.show()


    
```

```python
def dgp(array_size,start_gap,end_gap):
    
    start_gap = abs(start_gap)*-1
    end_gap = abs(end_gap)*-1

    gap_size = abs(start_gap - end_gap)

    #seasonality
    cycle = [np.sin(np.round(i)) for i in range(0,array_size)]

    #trend
    trend = np.linspace(0,10,array_size) + np.random.normal(size=array_size,scale=2)

    #Creating original series
    l1 = pd.Series([randint(0, 100) for i in range(array_size)] ) + trend + cycle

    #Creating and putting a hole in the second series
    l2 = l1.copy()
    l2[start_gap:end_gap] = [0 for i in range(0,gap_size)]
    
    return([l1,l2])
```

```python
def visual_analysis(array_size=18,start_gap=12,end_gap=3,**kwargs):
    l = dgp(array_size=array_size,start_gap=start_gap,end_gap=end_gap)
    df = auto_sarimax_v2(l[1], verbose=True)
    on_screen_number_one(df,l[1],size=len(l[0]))
    plt.show()
```

```python
def comparative_analysis(iterations,array_size,start_gap,end_gap):
    
    #empty dataframe
    pdf = pd.DataFrame()
    
    #for displaying whether the simulated data has an imposed gap or not
    series_type = ['No gap','Gap']
    
    # iterates through the number of simulations and the series type (no gap or gap)
    for i in range(0,iterations):
        print(f"Running {i}")
        
        #Generates data
        l = dgp(array_size=array_size,start_gap=start_gap,end_gap=end_gap)
        
        #Runs forecasts per series type per array that is generated in the dgp function
        for i,s in enumerate(series_type):
            
            try:
                #Runs forecast
                df = auto_sarimax_v2(l[i], verbose=True)

                #gathers the predicted values
                order = df[df.mse==min(df.mse)].order.to_list()[0]
                aic = df[df.mse==min(df.mse)].aic.to_list()[0]
                bic = df[df.mse==min(df.mse)].bic.to_list()[0]
                mse = df[df.mse==min(df.mse)].mse.to_list()[0]
                point = df[df.mse==min(df.mse)].forecast.to_list()[0][0]
                series = df[df.mse==min(df.mse)].series.to_list()[0]
                forecast = df[df.mse==min(df.mse)].forecast.to_list()[0]

                #appends to the dataframe
                pdf = pdf.append(pd.DataFrame({\
                  'series':[l[i]]
                  , 'series_type':[s]                             
                    ,'order':[order]\
                    ,'aic':[aic]\
                    ,'bic':[bic]\
                    ,'mse':[mse]\
                    ,'point':[point]
                    ,'series':[[series]]
                    ,'forecast':[forecast]\
                    }))
            except Exception as e:
                print(e)
                continue
            
    return(pdf.reset_index(drop=True))
            
```

```python
def dig(df,row):
    
    
    data = np.asarray(df.series.iloc[row][0])
    data_df = pd.DataFrame({'date':pd.date_range(datetime.date.today(), freq='M', periods=len(data))\
                       ,'data':data})\
                        .set_index('date')
    
    seasonal_orders = [df.order.iloc[row][1],(0, 0, 0, 12)]
    
    fdf = pd.DataFrame({'date':pd.date_range(datetime.date.today(), freq='M', periods=len(data)+6)\
                       ,'series':data.tolist()+[None for i in range(0,6)]
                       })
    
    
    for o in seasonal_orders:
    
        model = SARIMAX(data_df,order=(df.order.iloc[row][0]),seasonal_order=(o))
        model_fit = model.fit(disp=False)
        
        #print(model_fit.summary())
        fcast = model_fit.get_forecast(steps=6)
        print(f'\nThis is for {o}\n')
        print(f'AIC: {model_fit.aic}\nBIC: {model_fit.bic}\nMSE: {model_fit.mse}')
        print(fcast.summary_frame())
        fdf[f'forecast_{o}'] = [None for i in range(0,len(data))]+fcast.summary_frame()['mean'].to_list()
        
    fdf.set_index('date',inplace=True)
    fdf.plot()
    #plt.show()
    #return(fdf)
        
    
    
```

```python
class NotComformableError(Exception):
    # Initialize Class for a custom error
    def __init__(self, message):
        self.message = message
        
def fix_dates(df):
    
    """
    NAME: fix_dates
    PURPOSE: Puts the dates into the index of the dataframe
    ARGUMENTS:
        df(pandas dataframe) = dataframe to fix
    RETURNS:
        df(pandas dataframe) = dataframe with dates in index
    """
    
    # Detects the date column
    col = [c for c in d2.columns if isinstance(d2[c][0], pd.Timestamp)]
    
    # Fixes the dates to put them into the index
    df.set_index(col[0],inplace=True)
    
    return df
      
def dataframe_check(df):
    
    """
    NAME: dataframe_check
    PURPOSE: Checks to see if the dataframe has either an index of dates or a column of dates, throw error otherwise
    ARGUMENTS: 
        df(pandas dataframe) = dataframe to check
    RETURNS:
        None
    """
    
    msg = """This function requires a Pandas DataFrame that has ONE data column and requires either the index to be a DateTimeIndex or requires a column of that is a Timestamp."""
    
    if isinstance(df, pd.DataFrame):
        try:
            # Checks to see if the dataframe has an index of dates
            if type(df.index) == pd.DatetimeIndex:
                return df
            #Checks to see 
            elif df.shape[1] == 2 and  [c for c in d2.columns if isinstance(d2[c][0], pd.Timestamp)]:
                df = fix_dates(df)
                return df
            elif df.shape[1] > 2:
                raise NotComformableError(f"Looks like this dataframe has too many ({df.shape[1]}) columns. "+msg)
            #throw custom error
            else:
                raise NotComformableError()
        except Exception as e:
            print(e)
    else:
        raise NotComformableError("Looks like this isn't a dataframe. " + msg)

def STL_decomp(series,period,**kwargs):
    """
    NAME: STL_decomp
    PURPOSE: Creates a seasonal decomposition of the time series and sends a table to GBQ
    ARGUMENTS: 
        series(pandas dataframe with date as index) = time series 
    RETURNS:
        None
    """
    
    #Check and fix dataframes.  Throw error if not obvbiously conformable.  
    series = dataframe_check(series)
    
    #Seasonal decomposition using STL
    stl = STL(series.data, period=period).fit()
    
    d = {}
    
    # Allowing for updating the dictionary that goes into the dataframe with things like "academic partner"
    for k,v in kwargs.items():
        d.update({k:v})
    
    # Updating the dictionary with the STL data
    d.update({'actuals':stl.observed
            ,'trend':stl.trend\
            ,'seasonal':stl.seasonal\
            ,'residual':stl.resid})
    
    #putting together a DF
    df = pd.DataFrame(d).reset_index()
    
    return df

def send_STL_to_gbq(df):

    """
    NAME: send_STL_to_gbq
    PURPOSE: Sends a dataframe to GBQ
    ARGUMENTS:
        df(pandas dataframe) = dataframe
    RETURNS:
        None
    """
    # Send dataframe to GBQ

    from google.cloud import bigquery

    client = bigquery.Client()

    job_config = bigquery.LoadJobConfig(write_disposition='WRITE_APPEND')

    job = client.load_table_from_dataframe(df,destination='pols_all_pols_forecast_models.ref_stl_seasonal_decomposition',job_config=job_config)

    job.result()
```
